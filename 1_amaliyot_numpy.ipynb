```python
import numpy as np

raw = np.random.randint(40000, 200000, 200)
daromad = (raw // 500) * 500 # berilgan sonlarni uzbek kupyurasiga mantiqan tenglab oldik

ortacha=np.mean(daromad) # umumiy o'rta daromad
median=np.median(daromad) # 200 ta haydovchining haqiqiy markaziy daromadi
saralangan=np.sort(daromad)[::-1] # kattadan kichikka tartiblaymiz
top10=saralangan[:20] # eng yuqori daromadli 10%
top10_ortacha=np.mean(top10)
farq=top10_ortacha-median # top 10% haydovchining o'rtacha haydovchiga nisbatan daromad farqi
foiz = (farq / median) * 100 # top 10% haydovchi oddiy haydovchidan 60% ko'proq daromad qilyapti

print("\nðŸ“Š TAKSI HAYDOVCHILARI HISOBOTI")
print("-" * 40)

print(f"Haydovchilar soni: {len(daromad)}")
print(f"Oâ€˜rtacha daromad (mean): {ortacha:,.0f} soâ€˜m")
print(f"Median daromad: {median:,.0f} soâ€˜m")

print("\nðŸ”¥ TOP 10% eng kuchli haydovchilar")
print(f"Ularning oâ€˜rtacha daromadi: {top10_ortacha:,.0f} soâ€˜m")

print("\nðŸ“ˆ Taqqoslash")
print(f"Oddiy haydovchidan farqi: {farq:,.0f} soâ€˜m")
print(f"Foizda ustunligi: {foiz:.1f}%")

print("\nðŸ§  Biznes xulosa:")
if foiz > 50:
    print("TOP haydovchilar juda samarali. Ularga maxsus bonus tizimi kerak va ishlash uslublarini qolganlarga tadbiq etish kerak")
elif foiz > 20:
    print("TOP haydovchilar ancha yaxshi ishlayapti. Ragâ€˜batlantirish foydali boâ€˜ladi.")
else:
    print("Haydovchilar daromadi bir xil darajada. Tizim barqaror.")



```
